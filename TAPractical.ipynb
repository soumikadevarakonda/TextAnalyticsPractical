```python
#Experiment - 2

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.bbc.com/news"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
paragraphs = soup.find_all("h2")
i = 1

for paragraph in paragraphs[:10]:
    text = paragraph.get_text()
    if text:
      print(f"{i}. {text}")
      i += 1
```

```python
#Experiemnt - 3


import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
import re

sample_text = """
Text analytics is the process of analyzing text data.
Text analytics helps extract meaningful information from text.
Text visualization helps understand text analytics better.
"""

with open("sample.txt", "w") as file:
    file.write(sample_text)

with open("sample.txt", "r") as file:
    text = file.read().lower()

words = re.findall(r'\b\w+\b', text)

word_freq = Counter(words)

top_10 = word_freq.most_common(10)
words_list, freq_list = zip(*top_10)

print("Bar Chart\n")
plt.figure()
plt.bar(words_list, freq_list)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 10 Frequent Words")
plt.xticks(rotation=45)
plt.show()

print()
print()

print("Word Cloud\n")
wordcloud = WordCloud(width=800, height=400, background_color='white')
wordcloud.generate_from_frequencies(word_freq)

plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Word Cloud of Text Document")
plt.show()
```

```python
#Experiment - 4

import nltk
from nltk.corpus import gutenberg, brown

nltk.download('gutenberg')
nltk.download('brown')

print("File identifiers in Gutenberg corpus:\n")
for file_id in gutenberg.fileids():
    print(file_id)

print("\nGenres in Brown Corpus:\n")
print(brown.categories())

editorial_words = brown.words(categories='editorial')
editorial_word_types = sorted(set(editorial_words))

print("\nTotal unique words in editorial genre:", len(editorial_word_types))
print("\nFirst 20 sorted word types:")
print(editorial_word_types[:20])

selected_words = editorial_word_types[3493:3499]
print("\nSelected words:")
print(selected_words)

target_genres = ['fiction', 'lore', 'belles_lettres', 'government', 'learned']

print("\nWord Frequency across genres:\n")
for word in selected_words:
    print(f"Word: '{word}'")
    for genre in target_genres:
        count = brown.words(categories=genre).count(word)
        print(f"  {genre}: {count}")
    print("-" * 30)


```

```python
#Experiment - 4 (2)

import nltk
from nltk.corpus import inaugural
import matplotlib.pyplot as plt

# Download required corpus
nltk.download('inaugural')

# 1. Print all file identifiers
file_ids = inaugural.fileids()
print("Inaugural Corpus File Identifiers:")
print(file_ids)

# Lists to store data
years = []
president_counts = []
congress_counts = []

# 2. Count occurrences of words
for file_id in file_ids:
    year = int(file_id[:4])   # Extract year from filename
    words = [w.lower() for w in inaugural.words(file_id)]

    years.append(year)
    president_counts.append(words.count('president'))
    congress_counts.append(words.count('congress'))

# 3. Plot the graph
plt.figure(figsize=(12,6))
plt.plot(years, president_counts, label='president', marker='o')
plt.plot(years, congress_counts, label='congress', marker='s')
plt.xlabel('Year of Presidential Address')
plt.ylabel('Word Count')
plt.title('Frequency of "president" and "congress" in Inaugural Addresses')
plt.legend()
plt.grid(True)
plt.show()
```
